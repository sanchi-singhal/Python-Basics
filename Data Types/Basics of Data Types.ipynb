#### Numeric
Represents the data which has numeric value.<br>
1.<b>Integers:</b> contains positive or negative whole numbers, no limit to how an integer value can be.<br>
2.<b>Float:</b> a real number with floating point representation(by decimal point) or with scientific notation (using e)<br>
3.<b>Complex Number:</b>(real part)+(imaginary part)j.

#### Boolean
Represents one of the two values i.e., True or False.<br>
It is of <class 'bool'><br>
<b>bool([x]):</b> to return or convert a boolean value.

```python
print(bool(2==4))
print(bool(None))
print(bool({})) # for all empty data types it gives False
print(bool("hi"))
```

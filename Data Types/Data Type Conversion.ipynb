It is the process of converting the value of one data type to another data type.<br>
#### <u>Types</u>:
<b>1.Implicit</b>: the Python interpreter automatically converts one data type to another without any user involvement.(lower data type to higher data types to avoid data loss) <br>
Eg: int -> float -> complex number<br>
    When adding int and str it will give TypeError , here we use explicit type conversion

```python
a=5      #int
b=5.5    #float
print(a+b)    #float
```

```python
a=5   #int
b=5+2j  #complex
print(a+b)   #complex
```

#### 2.Explicit(Typecasting):
User converts the data type of variable of value of needed data type.<br>
There is a risk of data loss as we are forcing an expression to be changed in some specific data type.

bin(x): Converts an integer into binary string.

```python
print(int(5))
# print(int("shdh",2)) #can't converted to int not numeric
s="101010"
print(int(s,2))
print(float(4))
print(complex(2,3))
print(str([2,3,4]))
print(len(str([2,3,4])))
print(type(str([2,3,4])))
print(dict(((1,2),(3,4))))
print(eval("1+2"))  #solve expression
print(eval("2%4+3"))
x=3
print("Value:",eval("x+3*3"))  #expression must be in form of a string.
print(chr(64))
print(bin(34))
print(hex(34))
print(oct(34))
# print(unichr(34))   #not in python 3 versions
print(ord("A"))
```

### Difference between str() and repr()
1.str() is used for creating output for end user while repr() is mainly used for debugging and development. reprs goal is to be unambiguous and strs is to be readable. <br>
2.repr() compute the official string representation of an object (a representation that has all information about the object) and str() is used to compute the informal string representation of an object (a representation that is useful for printing the object).

```python
import datetime
today = datetime.datetime.now()
  
# Prints readable format for date-time object
print (str(today))
  
# prints the official format of date-time object
print (repr(today))  
```

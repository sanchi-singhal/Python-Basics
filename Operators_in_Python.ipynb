#### Arithmetic Operators
used with numeric values to perform common mathematical operations.


(x%10 gives last digit)<br>
(x%100 gives last two digits)<br>
//(integer division) is called floored division because if any number is negative then the output will be floored,it returns closest integer value which is less than or equal to the expression value. eg:5/2=2.5 so 5//2=2<br>
/ called float division because it always return float value

```python
print(4+2)     # can return both int and float value
print(3-4)     #can return both int and float value
print(4*2)     #can return both int and float value
print(-4*2)              
print(4/2)     # / always return float value
print(-5/2)    
print(4%2.0)   #can return both int and flaot value
print(-5%2)    #always results in positive
print(9.0//2.0) #can return both int and float value
print(9//2)
print(9//2.0)
print(9.0//2.0)
print(-9//2)      #if one of operand is negative result is floored.
```

#### Operators Precedence and Associativity

Operator Precedence:This is used in an expression with more than one operator with different precedence to determine which operation to perform first.<br>
Operator Associativity: If an expression contains two or more operators with the same precedence then Operator Associativity is used to determine. It can either be Left to Right or from Right to Left.<br>
Operators Precedence and Associativity are two main characteristics of operators that determine the evaluation order of sub-expressions in absence of brackets.

only ** has associativity from right to left . All other groups has associativity from left to right
